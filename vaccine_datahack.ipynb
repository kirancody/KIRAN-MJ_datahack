```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('training_set_features.csv')
```

```python
df.opinion_xyz_sick_from_vacc.dtype
df
```

```python
dfxyz=df[['health_worker','doctor_recc_xyz','behavioral_touch_face','behavioral_outside_home','behavioral_large_gatherings','behavioral_wash_hands','behavioral_face_mask','behavioral_avoidance','behavioral_antiviral_meds','chronic_med_condition','opinion_xyz_vacc_effective','opinion_xyz_risk']]
dfxyz['opinion_xyz_sick_from_vacc']=1/(1+df['opinion_xyz_sick_from_vacc'])
```

```python
dfxyz.head()
dfxyz.shape
```

```python
dfxyz=dfxyz.fillna(0)
```

```python
dfxyz.shape
```

```python
df_predict=pd.read_csv('training_set_labels.csv')
df_predict.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1=LogisticRegression()
```

```python
model1.fit(dfxyz,df_predict.xyz_vaccine)
```

```python
k=model1.predict(dfxyz)
np.sum(k)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(k,df_predict.xyz_vaccine)
```

```python
dfseas=df[['health_worker','doctor_recc_seasonal','behavioral_touch_face','behavioral_outside_home','behavioral_large_gatherings','behavioral_wash_hands','behavioral_face_mask','behavioral_avoidance','behavioral_antiviral_meds','chronic_med_condition','opinion_seas_vacc_effective','opinion_seas_risk']]
dfseas['opinion_seas_sick_from_vacc']=1/(1+df['opinion_seas_sick_from_vacc'])
```

```python
dfseas=dfseas.fillna(0)
```

```python
dfseas.head()
```

```python
dfseas.shape
```

```python
model2=LogisticRegression()
model2.fit(dfseas,df_predict.seasonal_vaccine)
```

```python
m=model2.predict_proba(dfseas)[:,1]
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(df_predict.seasonal_vaccine,m)
```
